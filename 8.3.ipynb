```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
POINTS=np.array([[0.2, 1.8], [-0.7,0.7], [-0.3,-1.8], [1.2,-1.6], [2.1,1.1], [-2.4,0.5]])
```

```python
A = np.power(POINTS,2)
L = np.matmul(A.T,A)
R = np.matmul(A.T, np.ones(len(POINTS)).T)
```

### Solve the system $ A^T A \vec{x} = A^T \vec{b} $

```python
alpha, beta = np.linalg.solve(L, R)  
a = np.sqrt(1/alpha)
b = np.sqrt(1/beta)
```

### Plot

```python
x = np.linspace(-a,a,200)
y = np.sqrt(b**2*(1-x**2/a**2))
fig, ax = plt.subplots(1,1)
for point in POINTS[:-1]:
    ax.scatter(*point, c = 'k')
ax.scatter(*POINTS[-1], c='k', label='observations')
ax.plot(x,y, c='b', label='predicted orbit')
ax.plot(x,-y, c='b')
ax.set(xlabel='x', ylabel='y', title='CERES orbit')
ax.legend(loc=(1.05,0.9))
```
