```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def u(k):
    if k<=1:
        return 2
    else:
        return 2**(k-1) * np.sqrt(2*  (1-np.sqrt(  1-(2**(-k+1)*u(k-1))**2))   )
```

```python
PI = np.pi
seq = np.array([u(i) for i in range(30)])
err = abs(seq-PI)
```

```python
min(err)
```

```python
plt.semilogy(err, marker= '.')
```
