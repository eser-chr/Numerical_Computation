```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# The main functions of Newton's method
def newton_step(x, f, df):
    return x-f(x)/df(x)

def newton(x_0, f, df, ITERATIONS=20):
    
    temp = x_0
    results = np.zeros(ITERATIONS)
    results[0]=x_0
    for i in range(ITERATIONS):    
        temp = newton_step(temp, f, df)
        results[i]=temp
    return results
```

```python
# The given functions and their derivatives

def f1(x):
    return x**2

def f2(x):
    return np.exp(x)-2

def f3(x):
    return abs(x)**1.5

def df1(x):
    return 2*x

def df2(x):
    return np.exp(x)

def df3(x):
    return 1.5*np.sqrt(x) if x>=0 else 1.5*np.sqrt(-x)

def f4(x):
    return 1/x - 1

def df4(x):
    return -1/x**2
```

```python
# The real solutions
f1_zero = 0
f2_zero =np.log(2) 
f3_zero = 0
```

```python
x = np.linspace(-1,1)
fig, ax = plt.subplots(1,1)
ax.plot(x, f1(x), label = 'f1')
ax.plot(x, f2(x), label = 'f2')
ax.plot(x, f3(x), label='f3')
ax.legend()
```

```python
x = np.linspace(-1,1)
fig, ax = plt.subplots(1,1)
ax.plot(x, df1(x), label = 'df1')
ax.plot(x, df2(x), label = 'df2')
ax.plot(x, np.array(list(map(df3, x))), label = 'df3')
ax.legend()
```

```python
f1_res = newton(0.5,f1,df1)
f2_res = newton(0.5,f2,df2)
f3_res = newton(0.5,f3,df3)
```

```python
ITERATIONS = np.array(list(range(1,20)))

fig, ax = plt.subplots(1,1)
ax.plot(f1_res)
ax.plot(f2_res-f2_zero)
ax.plot(f3_res)
ax.plot(ITERATIONS, 0.001/ITERATIONS**2)

ax.set(yscale='log')
```
