```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def x(n, perturbation):
    if n<=0:
        return 1+perturbation
    else:
        return x(n-1, perturbation)/3
```

```python
def y(n, perturbation_0, perturbation_1):
    if n<=0:
        return 1+perturbation_0
    elif n==1:
        return 1/3+perturbation_1
    else:
        return 4*y(n-1, perturbation_0, perturbation_1)/3 - y(n-2, perturbation_0, perturbation_1)/3
```

```python
def z(n, perturbation_0, perturbation_1):
    if n<=0:
        return 1+perturbation_0
    elif n==1:
        return 1/3+perturbation_1
    else: 
        return 10*z(n-1, perturbation_0, perturbation_1)/3 - z(n-2, perturbation_0, perturbation_1)
```

```python
N = 20
REAL = np.array([(1/3)**i for i in range(N)])
x_val = np.array([x(i,0) for i in range(N)])
y_val = np.array([y(i,0,0) for i in range(N)])
z_val = np.array([z(i,0,0) for i in range(N)])
```

```python
x_err=abs(x_val-REAL)
y_err=abs(y_val-REAL)
z_err=abs(z_val-REAL)
```

```python
fig, ax = plt.subplots(1,1)
ax.plot(x_err, label = 'x')
ax.plot(y_err, label = 'y')
ax.plot(z_err, label = 'z')
ax.set(yscale='log')
ax.legend()
```

# PERTURBED

```python
N = 20
REAL = np.array([(1/3)**i for i in range(N)])
x_val = np.array([x(i,1e-14) for i in range(N)])
y_val = np.array([y(i,1e-14,1e-14) for i in range(N)])
z_val = np.array([z(i,1e-14,1e-14) for i in range(N)])
```

```python
x_err=abs(x_val-REAL)
y_err=abs(y_val-REAL)
z_err=abs(z_val-REAL)
```
