```python
from scipy.sparse.linalg import gmres
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 1000
A = np.eye(n)
A[:-1, 1:] += np.eye(n-1)
b = np.ones(n)
```

```python
errors=np.zeros(20)
print(errors)
for i in range(20):
    sol = gmres(A, b, maxiter=10*i)[0]
    res = np.matmul(A,sol)-b
    errors[i] = np.linalg.norm(res)
```

```python
errors
```

```python
plt.semilogy(errors)
```
