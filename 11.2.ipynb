```python
from scipy.sparse.linalg import gmres
import numpy as np
import matplotlib.pyplot as plt
from time import time
```

```python
n = 1000
A = np.eye(n)
A[:-1, 1:] += np.eye(n-1)
b = np.ones(n)
```

```python
N = 100
errors=np.zeros(N)
for i in range(N):
    sol = gmres(A, b, maxiter=i, restart=i+1)[0]
    res = A@sol-b
    errors[i] = np.linalg.norm(res)
```

```python
errors
```

```python
fig, ax = plt.subplots(1,1)
ax.plot(errors)
ax.set(yscale = 'log',
    title = 'GMRES | Residual convergence', xlabel = 'iterations', ylabel = 'residual')
ax.grid()
```
