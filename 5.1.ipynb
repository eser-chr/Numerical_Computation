```python
import numpy as np
import matplotlib.pyplot as plt
from typing import Callable
```

```python
def Q(f:Callable, N:int):
    h = 1/(N**0.5)
    total = 0
    for i in range(-N, N+1):
        total+=f(i*h)
    return total*h
```

```python
def f(x):
    return np.exp(-x**2)*(np.sin(x))**2
```

```python
x = np.linspace(-4,4,200)
plt.plot(x, f(x))
plt.xlabel('x')
plt.ylabel('f(x)')
```

```python
Q(f, 100)
```

```python
N = np.logspace(0,5, 20)
real = Q(f, int(1e6))
errors = [abs(Q(f, int(n))-real) for n in N]
```

```python
fig, ax = plt.subplots(1,1)
ax.plot(N, errors, marker='.')
ax.set(xscale='log', yscale='log', xlabel='N', ylabel='err', 
       title='f(x) = exp(-x^2)sin^2(x), trapezoidal rule error', ylim=[0,0.1])
```
