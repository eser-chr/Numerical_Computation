```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# The function and its derivative
def f(x):
    return x**2+3*x-4

def df(x):
    return 2*x+3

def Phi(x):
    return x**2+4*x-4
```

```python
# Plot and the point we want to approach using Newtons method
x = np.linspace(-5,5)
y = f(x)
plt.plot(x,y)
plt.scatter(1,f(1), c='r')
```

For $\Phi(x) = x^2+4x-4$ we can see that 
$|\Phi'(x^*)|=6>1$

Which implies that we have no contraction


```python
def fixed_point(x_0, Phi, ITERATIONS):
    temp = x_0
    res = np.zeros(ITERATIONS+1)
    res[0] = x_0
    for i in range(1,ITERATIONS+1):
        temp = Phi(temp)
        res[i] = temp
    return res
```

```python
res  = fixed_point(0, Phi, 4)
```

```python
plt.plot(res)
```

```python
def newton_step(x, f, df):
    return x-f(x)/df(x)

def newton(x_0, f, df, ITERATIONS=20):
    
    temp = x_0
    results = np.zeros(ITERATIONS)
    results[0]=x_0
    for i in range(1,ITERATIONS):    
        temp = newton_step(temp, f, df)
        results[i]=temp
    return results
```

```python
res = newton(0, f, df)
```

```python
fig, ax = plt.subplots(1,1)
ax.plot(res)
ax.set(ylabel='x_n', xlabel='n_th. iteration')
```

Draw of the first iterations

```python
xs = res[:4]
ys = [f(x) for x in xs]

x = np.linspace(-5,5)
y = f(x)
```

```python
fig, ax = plt.subplots(1,1)
ax.plot(x, y, c='grey', linestyle='--')
for i in range(len(xs)):
    ax.scatter(xs[i], ys[i], label = f'{i} - iteration', marker='o')

ax.set(xlabel = 'x', ylabel='y', title = 'First 3 iterations')
ax.legend(loc=(1.05, 0.1))
```
